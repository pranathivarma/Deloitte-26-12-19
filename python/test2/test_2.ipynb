```python
import pandas as pd
```

```python

l= pd.read_csv('results.csv')
```

```python
print("data of the matches played in Glasgow and the home team lost is: ")

l[(l.tournament == 'FIFA World Cup qualification') & (l.city == 'Glasgow') & (l.home_score==0)]

#data of the matches played in Glasgow and the home team lost is:
```

```python
print (" data of matches played in London where home team lost by > 4 goals: ")

l[(l.city == 'London') & (l.away_score - l.home_score >4)]

#data of matches played in London where home team lost by > 4 goals 
```

```python
#Statistics of homescreen in all matches:

l.agg({'home_score': ['min', 'max', 'mean', 'median']})
```

```python
# matches played in Vietnam grouped by home_team and data being filtered on home_score in descending

print ( " Matches played in Vietnam grouped by home_team and data being filtered on home_score in descending: ")
l[l.country == 'Vietnam'].groupby('home_team').filter(lambda g: len(g) > 2).groupby('home_score').size().sort_values(ascending=False)
```

```python
#Fetch the host country and year of FIFA WorldCup into a dictionary
l['year'] = pd.DatetimeIndex(l['date']).year
fifa = l[l['tournament']=='FIFA World Cup']
dates = fifa.date.tolist()
years=[]
for date in dates:
    year=date.split('-')[0]
    if year not in years:
        years.append(year)
dictionaryOfCountries={}
for i in range(len(years)):
    for j in range(len(fifa[fifa['year']==eval(years[i])].country.unique())):
        if years[i] in dictionaryOfCountries.keys():
            dictionaryOfCountries[years[i]].append(fifa[fifa['year']==eval(years[i])].country.unique()[j])
        else:
            dictionaryOfCountries[years[i]]=[fifa[fifa['year']==eval(years[i])].country.unique()[j]]

dictionaryOfCountries
```
